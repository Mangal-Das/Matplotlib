```python
# importing libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('weatherHistory.csv')


x_train=df['Visibility (km)'].head(50).values
y_train=df['Wind Speed (km/h)'].head(50).values
```

```python
# printing dataset
print(x_train)
print(y_train)
```

```python
plt.scatter(x_train,y_train)
w,b=np.polyfit(x_train,y_train,1)
plt.plot(x_train,w*x_train+b)
```

```python
def cost_function(w,x,y):
    m=x_train.shape[0]
    for i in range(m):
        f_wb=x.dot(w)
        cost=sum((f_wb-y)**2)/(2*m)
    return cost

w1=np.linspace(1,4,20)
total_cost=[cost_function(w,x_train,y_train)for w in w1]

plt.plot(w1,total_cost,label="Cost Function")
plt.title("Grph of Cost Function")
plt.xlabel("W")
plt.ylabel("Total Cost")
plt.legend()
plt.grid(True)
plt.show()


```

```python
import math
num=math.e
def cost_function(w,x,y,b):
    m=x_train.shape[0]
    for i in range(m):
        f_wb=x.dot(w)+b
        cost=np.sum((f_wb-y)**2)/(2*m)
    return cost

def gradient_descent(x,y,alpha,iteration):
    w=3.5
    b=0
    m=x.shape[0]
    
    for _ in range(iteration):
        f_wb=x.dot(w)+b
        dj_dw=(np.sum((f_wb - y)*x))/m
        dj_db=(np.sum(f_wb - y))/m
        w=w-alpha*dj_dw
        b=b-alpha*dj_db
        cost=cost_function(w,x_train,y_train,b)

        
        print("iteration : ", _ , "cost :",cost,"w : ",w,"b :",b)
        
        
            
    return w,b,cost
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values for slope (w) and y-intercept (b)
w = -3.6
b = -2.8

# Sample data for demonstration
np.random.seed(0)
x_train = 2 * np.random.rand(100, 1)
y_train = 4 + 3 * x_train + np.random.randn(100, 1)

# Compute the predicted values using the given slope (w) and y-intercept (b)
f_wb = w * x_train + b

# Sort the x_train and f_wb based on x_train
sorted_indices = np.argsort(x_train[:, 0])
x_train_sorted = x_train[sorted_indices]
f_wb_sorted = f_wb[sorted_indices]

# Plot the best-fit line
plt.plot(x_train_sorted, f_wb_sorted, color='red', label='Best Fit Line')
plt.scatter(x_train, y_train, label='Data Points')
plt.xlabel('X_train')
plt.ylabel('Y_train')
plt.legend()
plt.grid(True)
plt.title('Best Fit Line for Linear Regression')
plt.show()


    
```
