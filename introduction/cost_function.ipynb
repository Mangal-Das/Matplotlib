```python
# importing libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('weatherHistory.csv')


x_train=df['Visibility (km)'].head(20).values
y_train=df['Wind Speed (km/h)'].head(20).values
```

```python
# printing dataset
print(x_train)
print(y_train)
```

```python
plt.scatter(x_train,y_train)
w,b=np.polyfit(x_train,y_train,1)
plt.plot(x_train,w*x_train+b)
```

```python
def cost_function(w,x,y):
    m=x_train.shape[0]
    for i in range(m):
        f_wb=x.dot(w)
        cost=sum((f_wb-y)**2)/(2*m)
    return cost

w1=np.linspace(1,4,20)
total_cost=[cost_function(w,x_train,y_train)for w in w1]

plt.plot(w1,total_cost,label="Cost Function")
plt.title("Grph of Cost Function")
plt.xlabel("W")
plt.ylabel("Total Cost")
plt.legend()
plt.grid(True)
plt.show()


```
