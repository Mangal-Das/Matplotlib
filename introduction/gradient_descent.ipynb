```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1,4,3,9,7,6,5,8,6,7])
y_train=np.array([1,3,6,4,5,2,7,9,8,10])
```

```python
plt.scatter(x_train,y_train)
```

```python
def cost_function(x,y,w,b):
    m=x_train.shape[0]
    f_wb=x.dot(w)+b
    cost=(np.sum((f_wb-y)**2))/(2*m)
    return cost
    
    
```

```python
cost_function(x_train,y_train,1,1)
```

```python
def gradient_descent(x,y,alpha,interation):
    new_w=[]
    j_wb=[]
    
    w= 0.56
    b=2.6
    m=x_train.shape[0]
    for _ in range (interation):
        f_wb=x.dot(w)+b
        dj_dw=(np.sum((f_wb-y)*x))/m
        dj_db=(np.sum((f_wb-y)))/m
        w=w-(alpha*dj_dw)
        
        b=b-(alpha*dj_db)
        new_w.append(w)
        
        cost=cost_function(x_train,y_train,w,b)
        j_wb.append(cost)
        print("Iteration :",_,"cost :",cost,"w:",w,"b:",b)
    return new_w,j_wb
        
        
    

    
```

```python
gradient_descent(x_train,y_train,0.01,10)
x,y=gradient_descent(x_train,y_train,0.01,10)
plt.plot(x,y)
```

```python
w=0.53
b=2.6
f_wb=x_train.dot(w)+b

plt.plot(x_train,f_wb,c='r',label="Regression Line")
plt.scatter(x_train,y_train,label="Experimental Data")
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
