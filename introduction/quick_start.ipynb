```python
import numpy as np
x=[]
y=[]
for i in range (1,10):
    x.append(i)

for j in range (10,19):
    y.append(j)

fig,ax=plt.subplots()
ax.set_title("Scatter Plot")
ax.set_xlabel("x")
ax.set_ylabel("Y")
ax.scatter(x,y)

print(x)
print(y)

```
