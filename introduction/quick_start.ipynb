```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (replace this with your dataset)
x_train = np.array([1, 2, 3, 4, 5])
y_train = np.array([2, 3, 2.5, 4, 3.5])

w = 1
b = 1.5

def calculate_graph(w, x, b):
    m = x.shape[0]
    f_wb = []
    for i in range(m):
        f = w * x[i] + b
        f_wb.append(f)
    return np.array(f_wb)

y_predicted = calculate_graph(w, x_train, b)

# Plot the data points
plt.scatter(x_train, y_train, label='Data Points')

# Plot the predicted line
plt.plot(x_train, y_predicted, color='red', label='Predicted Line')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('weatherHistory.csv',usecols=['Temperature (C)','Humidity'])
x_train=np.array(df['Temperature (C)'].head(100).values)
y_train=np.array(df['Humidity'].head(100).values)
m=x_train.shape[0]
w=0.4
b=1.7999999999999998
def calculate_graph(w,x,b):
    m=x.shape[0]
    f_wb=[]
    for i in range(m):
        f=w*x[i]+b
        f_wb.append(f)
    return np.array(f_wb)
y=calculate_graph(w,x_train,b)
plt.plot(x_train,y,c='r',label="Predicted Line")
plt.scatter(x_train,y_train,c='b',label="Given Data")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.legend()
plt.show()
```

```python
x_train=np.array([1,2,3,4,5])
y_train=np.array([1,3,5,4,6])
plt.scatter(x_train,y_train)
```

```python
w=1.3
b=0

def calc(w,x,b):
    m=x.shape[0]
    f_wb=[]
    for i in range(m):
        f=w*x[i]+b
        f_wb.append(f)
    return np.array(f_wb)
plt.plot(x_train,calc(w,x_train,b),c='b',label="Predicted Line")
plt.scatter(x_train,y_train,c='r',label="Given Data")
plt.legend()
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("My Practice Graph")
plt.show()
```

```python
# cost function 

def cost_function(w,b,x,y):
    
    m=x_train.shape[0]
    error =0
    for i in range(m):
        f_wb=w*x[i]+b
        error=error+(f_wb-y[i])**2
    total_error=error * (1/(2*m))
    return total_error
```
